```python
# Import required packages
import csv
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns
from mpl_toolkits.axes_grid1.inset_locator import mark_inset
```

```python
filepath=['csv/HF_1s_OUT','csv/HF_1s_KS_OUT','sto/HF_STO6G_OUT','sto/HF_STO3G_OUT',\
          'csv/XA_1s_OUT','csv/XA_1s_KS_OUT','sto/XA_STO6G_OUT','sto/XA_STO3G_OUT']

bond_length = np.linspace(0.5,3,26) 

# Define all the possible array that need to be read from the csv files  
E = []

for idx, path in enumerate(filepath):
    E.append([])
    
    # Open the CSV file in read mode
    filename = f"{path}.csv"
    
    with open(filename, 'r') as file:
        reader = csv.reader(file)

        # Skip the header row
        next(reader)

        # Read the values from each row
        for row in reader:
            if row:  # Check if the row is not empty
                try:
                    E_value = float(row[-1])  # Convert the last column's value to a float
                    E[idx].append(E_value)
                except ValueError:
                    print(f"Warning: Couldn't convert '{row[-1]}' to float.")
```

```python
# Apply the default Seaborn theme
sns.set_theme()

# Get the Seaborn color palette
colors = sns.color_palette("tab10",10)

fig, ax1 = plt.subplots(figsize=(12, 12), facecolor='white')

ax1.set(xlim=(0.3, 3.2), ylim=(-1.1,-0.2))

# Original plots with Seaborn colors
ax1.plot(bond_length, E[0], color=colors[0], linewidth=5, label="HFS/1$s$", linestyle=(0,(0.1,1.5)), dash_capstyle = 'round')
ax1.plot(bond_length, E[2], color=colors[1], linewidth=3, label="HFS/STO-6G")
ax1.plot(bond_length, E[3], color=colors[2], linewidth=3, label="HFS/STO-3G")

ax1.set_xlabel("Bond length (bohr)", fontsize=26)
ax1.set_ylabel(r'Total energy H$_2$ (hartree)', fontsize=26)

# Add legend to the main plot with manually set location and white background
ax1.legend(loc='upper left', bbox_to_anchor=(0.08, 0.96), fontsize=18, facecolor='white')

plt.xticks(fontsize=20)
plt.yticks(fontsize=20)

# Manually adding the inset plot at [x, y, width, height]
axins = fig.add_axes([0.49, 0.45, 0.5, 0.4])

# Set background color for inset plot to white
axins.set_facecolor('white')

# Add grey axis lines
for spine in axins.spines.values():
    spine.set_edgecolor('grey')

# Add grey grid
axins.grid(True, linestyle='--', linewidth=0.5, color='grey')

# Add dashed vertical line at x=1.4 in the inset plot
axins.axvline(x=1.4, color='r', linestyle='--', linewidth=1.5)

# Inset scatter plots with Seaborn colors
axins.scatter(bond_length, E[0], color=colors[0], label="HFS/1$s$", marker='o', s=90)
axins.scatter(bond_length, E[2], color=colors[1], label="HFS/STO-6G", marker='s', s=43)
axins.scatter(bond_length, E[3], color=colors[2], label="HFS/STO-3G", marker='s', s=45)

# Setting the area to be zoomed into
axins.set_xlim(0.95, 2.05)
axins.set_ylim(-1.07, -0.97)

# Add rectangle and lines indicating the zoom
pp, p1, p2 = mark_inset(ax1, axins, loc1=2, loc2=4, fc="none", ec="black")

# Set the line width of the connecting lines and the rectangle
pp.set_linewidth(0.5)
p1.set_linewidth(0.5)
p2.set_linewidth(0.5)

# Legend and ticks for inset
axins.legend(loc='upper right', bbox_to_anchor=(0.9, 0.3), fontsize=16)
axins.tick_params(axis='x', labelsize=18)
axins.tick_params(axis='y', labelsize=18)

plt.show()
```
